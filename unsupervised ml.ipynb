# unsupervised machine learning

```python
# unsupervised machine learning  (used bu iris.csv)
# Model khud hi data ke pattern, relation aur structure ko samajhta hai bina kisi predefined output ke.
 # K-Means Clustring:-
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Prince Kumar\Downloads\iris_dataset.csv")
```

```python
dataset.head(3)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
x = dataset.iloc[:, :-1]   # last column (Species) hata diya
```

```python
x = dataset.drop('target', axis=1)
```

```python
wcss = []

for i in range(2, 21):
    km = KMeans(n_clusters=i, init='k-means++', random_state=42)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
plt.plot([i for i in range(2,21)],wcss)
plt.xlabel("no of cluster")
plt.ylabel("wcss")
plt.show()
```

```python
plt.plot([i for i in range(2,21)],wcss)
plt.xlabel("no of cluster")
plt.xticks([i for i in range(2,21)])
plt.ylabel("wcss")
plt.show()
```

```python
plt.figure(figsize=(12,6))   #or accha show ho graph
plt.plot([i for i in range(2,21)],wcss,marker="o")
plt.xlabel("no of cluster")
plt.xticks([i for i in range(2,21)])
plt.ylabel("wcss")
plt.grid(axis = "x")
plt.show()
```

```python
from sklearn.cluster import KMeans

# step 1: sirf numerical data lo (last column ko hata do)
x = dataset.iloc[:, :-1]

# step 2: KMeans lagao
kmn = KMeans(n_clusters=3, init='k-means++', random_state=42)
dataset["Predict"] = kmn.fit_predict(x)

# step 3: output dekho
print(dataset.head())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dataset['target'] = le.fit_transform(dataset['target'])

kmn = KMeans(n_clusters=3, random_state=42)
dataset["Predict"] = kmn.fit_predict(dataset)
```

```python
sns.pairplot(data=dataset,hue="Predict")
plt.savefig("predict.jpg")
plt.show()
```
